```python
##this is for ngram, if nounphrase, need to get ngram 
import re, utils, math
from datetime import datetime
start = datetime.now()
tf_corpus=utils.open_pickle('txt train tf new nounphrase corpus')
candidates=utils.open_pickle('txt nounphrase candidates')

#tf_corpus=[[('milkshake',6), ('banana', 4),('chocolate',3),('chocolate milkshake', 2), ('milkshake banana', 3),
#             ('chocolate milkshake banana', 2),('machine',5)],
#           [('chocolate',7),('white chocolate', 4),('white', 3)],
#           [('machine',6),('learning', 6),('machine learning', 6)]]

#candidates=[[ ('milkshake banana', 3), ('banana', 4),('chocolate',3),('chocolate milkshake', 2), ('milkshake',6),
#             ('chocolate milkshake banana', 2),('machine',5)],
#           [('white chocolate', 4),('chocolate',7),('white', 3)],
#           [('machine learning', 6),('machine',6),('learning', 6)]]

##this is for general purpose, for ngram and noun phrase
#GDC and DPM
def calculate_statistical(tf_corpus):
    dpm_values = []
    for n_doc in candidates:
        doc = {}
        cand_perdoc = [x[0] for x in n_doc]
        for n_cand, value in n_doc:
            ##calculate DPM
            matches = []
            term_tf = value
            for n_cand2 in cand_perdoc:
                if re.findall(r'.*'+n_cand+'$', n_cand2):
                    matches.append((n_cand2))
            #only superterm
            tf_superterm_values = [x[1] for x in n_doc if x[0] in matches and x[0] != n_cand]
            if len(tf_superterm_values) == 0:
                tf_superterm_values = [0]
            #calculate the DPM index        
            dpm = float("{0:.5F}".format(1 - max (s_tf / term_tf for s_tf in tf_superterm_values)))
            
            ##calculate GDC
            length_term = len(n_cand.split(" ")) #GDC
            if length_term == 1:
                # 10% from gdc value for unigram term
                gdc = float("{0:.5F}".format( 0.1 * ( length_term * math.log10 (term_tf) * term_tf ) / term_tf ))  
            else:
                matched_word = [word for word in [word for word in n_cand.split(" ")] 
                            if word in [value[0] for value in n_doc]]
                freq_word = sum([value[1] for value in n_doc if value[0] in matched_word])
                gdc = float("{0:.5F}".format(( length_term * math.log10 (term_tf) * term_tf ) / freq_word ))
            #check if GDC value is inf
            if gdc == float("-inf"):
                gdc = 0.0
            doc[n_cand] = ((dpm, gdc))
        dpm_values.append(doc)
    return dpm_values
#save in pickle
dpm_values = calculate_statistical(tf_corpus)

def feature_statistical(candidates, dpm_values):
    feature = []
    for n_doc in range(len(candidates)):
        doc = []
        for n_cand in candidates[n_doc]:
            for term, value in dpm_values[n_doc].items():
                if n_cand[0] == term:
                    doc.append(value)
        feature.append(doc)
    return feature
            
print(feature_statistical(candidates, dpm_values))
print("time=", datetime.now()-start)
```

```python
#match value from topic rank into candidates
import re

topic=[['table','zoo','machine learning'], ['fried', 'chicken katsu']]
candidates=[['chocolate', 'chicken nugget', 'table','zoo','machine learning', 'dinner table', 'colchester zoo'],
            ['katsu', 'domino', 'fried', 'chicken katsu', 'fried frice']]
al=[]
for n_doc in range(len(candidates)):
    doc=[]
    matches=[]
    topics=topic[n_doc]
    for n_cand in candidates[n_doc]:
        for top in topics:
            if re.findall(r'.*'+top+'.*', n_cand):
                matches.append(n_cand)
        if n_cand in matches:
            doc.append(1)
        else:
            doc.append(0)
    al.append(doc)
    
print(al)
```
