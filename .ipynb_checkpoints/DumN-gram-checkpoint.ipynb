```python
import glob, os
from pandas import DataFrame
import pandas as pd

def concat(file):
    content=open(file,encoding='utf-8')
    con=' '.join(x.strip() for x in content)
    return con

#train='./semeval/train/*.txt.final'
train='./smalldummy/train/*.txt.final'

text=[(os.path.basename(files), concat(files)) for files in glob.glob(train)]

labels=['file', 'text']
df=DataFrame.from_records(text, columns=labels)
df['file']=df['file'].apply(lambda x: x.rstrip('.txt.final'))

print(df.head())
```

```python
#clean and extract title and some section
import re

def concat(file):
    content=open(file,encoding='utf-8')
    #certain part for features
    section=['abstract', 'introduction', 'methodology','conclusion']
    pat=re.compile(r'%s' %  '|'.join(section), re.IGNORECASE)
    
    sp = [x for x  in re.split(section, content) if x]
    dic = {k:v for k,v in zip(sp[0::2],sp[1::2])}
    return dic
    # filter unwanted string and character 
    
    con=' '.join(x.strip() for x in content)
    # remove number and [], ()
    return con       

te=[]
for files in glob.glob(train):
    te.append(os.path.basename(files))
    te.append(concat(files))
print(te)            
```

```python
#extracting the label from gold standard
import re
gold_std=open('./smalldummy/train/train.combined.final', encoding='utf-8').read()

all_keys=[]
for line in gold_std.split('\n'):
    actual_keys=[] 
    for key in line[7:].split(","):
        actual_keys.append(key)
    all_keys.append(actual_keys)

keys=pd.Series(all_keys)
df.insert(loc=2, column='keyphrases', value=all_keys)
print(df.head())
```

```python
from nltk import word_tokenize
from nltk import re
import string

df.text=df.text.apply(lambda x: x.lower())
df['token']=df['text'].apply(lambda x:''.join([i for i in x if i not in string.punctuation]))
#df['token']=df['token'].apply(lambda x:''.join([i for i in x if i not in string.digit]))
df['token']=df.apply(lambda x: word_tokenize(x['token']), axis=1)
#df['token']=df.apply(lambda x:''.join([word.lower() for word in word_tokenize(x['text']) if re.search("\w", word)]), axis=1)

print(df['token'])
```

```python
from nltk.util import ngrams
# put all preprocessing in here and clean unwanted characters
def extract_ngrams(text):
    n_grams=ngrams(word_tokenize(text),1)
    ngrams_list=[','.join(gram) for gram in n_grams]
    n_grams=ngrams(word_tokenize(text),2)
    ngrams_list2=[','.join(gram) for gram in n_grams]
    all_list=ngrams_list+ngrams_list2
    '''
     for i in range(5):
        ngrams=ngrams(text,i)
        all_ngrams=[','.join(x) for x in ngrams]
    '''
    return all_list

df['ngram']=df.apply(lambda x: extract_ngrams(x['text']), axis=1)
print(df['ngram'].head().tolist())
```
